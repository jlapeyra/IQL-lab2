```python
import requests
import re
import os
import sys
from os import listdir
from os.path import isfile, isdir, join
import string
from math import log, sqrt
import numpy as np
import random
```

```python
random.seed('111')
```

```python
S = requests.Session()
```

```python
URL = "https://tr.wikipedia.org/w/api.php"
PARAMS = {
    "action": "query",
    "format": "json",
    "list": "allcategories",
    "acmin": 100,
    "aclimit": 500
}
```

```python
f = open("wiki_categories.txt",'w')
for i in range(100):
    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()

    CATEGORIES = DATA["query"]["allcategories"]

    for cat in CATEGORIES:
        cat_name = cat["*"]
        m = re.search("[0-9]{4}",cat_name)
        n = 'taslakları' in cat_name
        if not m and not n:
            f.write(cat_name+'\n')

    if "continue" in DATA:
        PARAMS["acfrom"] = DATA["continue"]["accontinue"]
    else:
        break

f.close()
```

```python
with open('wiki_categories.txt', 'r') as data:
    categories = data.readlines()
    categories = random.sample(categories, 300)
```

```python
if not os.path.isdir("./categories"):
  os.mkdir("categories")
```

```python
if not os.path.isdir("./categories"):
  os.mkdir("categories")
for cat in categories:
    cat = cat.replace("\ufeff", "")
    cat_dir = "categories/"+cat.replace(' ','_')
    if not os.path.isdir(cat_dir):
        os.mkdir(cat_dir)
    title_file = open(os.path.join(cat_dir,"titles.txt"),'w')

    PARAMS = {
        "action": "query",
        "list": "categorymembers",
        "format": "json",
        "cmtitle": "Category:"+cat,
        "cmlimit": "100"
    }

    R = S.get(url=URL, params=PARAMS)
    DATA = R.json()

    PAGES = DATA["query"]["categorymembers"]

    for page in PAGES:
        title = page["title"]
        ID = str(page["pageid"])
        if title[:9] != "Kategori:":
            title_file.write(ID+' '+title+'\n')

    if "continue" in DATA:
        PARAMS["cmcontinue"] = DATA["continue"]["cmcontinue"]
    else:
        break

    title_file.close()
```

```python
def read_titles(filename):
    IDs = []
    titles = []
    f = open(filename,'r')
    for l in f:
        l.rstrip('\n')
        IDs.append(l.split()[0])
        titles.append(' '.join(l.split()[1:]))
    return IDs, titles
```

```python
j = 0
for cat in categories:
    cat = cat.replace("\ufeff", "")
    print("Processing category",cat)
    cat_dir = "categories/"+cat.replace(' ','_')
    title_file = os.path.join(cat_dir,"titles.txt")
    IDs, titles = read_titles(title_file)

    content_file = open(os.path.join(cat_dir,"linear.txt"),'w')
    
    for i in range(len(titles)):
        j += 1
        print(cat, j)
        PARAMS = {
            "action": "query",
            "prop": "extracts",
            "format": "json",
            "explaintext": True,
            "redirects": True,
            "titles": titles[i]
        }

        R = S.get(url=URL, params=PARAMS)
        DATA = R.json()

        PAGES = DATA["query"]["pages"]

        for page in PAGES:
            extract = PAGES[page]["extract"]
            content_file.write("\n")
            content_file.write(extract+'\n')
            content_file.write("\n\n")

    content_file.close()
```
